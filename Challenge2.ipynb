```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("D:/Recommender System Datasets/challenge2/recommendation challenge (Music)/train.csv/train.csv")
df.head()
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train.isnull().sum()
```

```python
train.shape
```

```python
data = train[:10000]
```

```python
piv = data.pivot_table(index='msno', columns='song_id', values='target', fill_value=0.5)
```

```python
piv
```

```python
piv.shape
```

```python
df = pd.DataFrame(piv)
```

```python
df.head()
```

```python
ulist = df.index.tolist()
slist = df.columns.tolist()
```

```python
A = np.array(df)
A
```

```python
from sklearn.decomposition import TruncatedSVD


k = 10


svd = TruncatedSVD(n_components=k)

svd.fit(A)

user_factors = svd.transform(A)
item_factors = svd.components_
```

```python
user_factors
```

```python
item_factors
```

```python
items = np.transpose(item_factors)
```

```python
len(item_factors[0])
```

```python
def rec(user):
    ix = ulist.index(user)
    user_latent = user_factors[ix]
    recs = []
    c=0
    for i in items:
        tar = np.dot(user_latent, i)
        recs.append([slist[c],tar])
        c=c+1
    return recs
```

```python
l = rec('FGtllVqz18RPiwJj/edr2gV78zirAiY/9SmYvia+kCg=')
l.sort(key = lambda x: x[1], reverse=True)
l
```

```python
def evaluation(user):
    l = rec(user)
    c=0
    td = test[test['msno']==user]
    ts = td['song_id'].tolist()
    tp=0
    fn=0
    tn=0
    fp=0    
    rs = []
    for i in l:
        if(i[0] in ts):
            
            ta = td[td['song_id']==i[0]]['target'].tolist()[0]
            rs.append([i[0],i[1], ta])
            
            if(i[1]>=0.5 and ta<0.5):
                fn=fn+1
            if(i[1]<0.5 and ta>=0.5):
                fp=fp+1
            if(i[1]>=0.5 and ta>=0.5):
                tp=tp+1
            if(i[1]<0.5 and ta<0.5):
                tn=tn+1

        c=c+1
    print("precision",tp/(tp+fp))
    print("Accuracy",(tn+tp)/(tn+tp+fn+fp))
    return rs
```

```python
evaluation('FGtllVqz18RPiwJj/edr2gV78zirAiY/9SmYvia+kCg=')
```
